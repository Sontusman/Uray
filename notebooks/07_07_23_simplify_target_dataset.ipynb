```python
import numpy as np
import pandas as pd
```

```python
sillyfilepath = "../data/raw/19_06_23_targets_chm51-1147.xlsx"
```

```python
def get_norm_target_data(sillyfilepath):
    raw_data = pd.read_excel(sillyfilepath, header = 2)
    new_data = raw_data.drop(columns  = ['BAR-CODE', 'comment1', 'comment2', 'comment3', 'Gender', 'Age'])
    new_data.columns = ["label", "ph", "ph_a>6.5", "ph_a>7", "sg", "sg_a", "pro", "pro_a>0.066", "pro_a>0.033", "pro_a>0", "glu", "glu_a", "ket", "ket_a", "bil", "bil_a", "urg", "urg_a", "mucus", "mucus_a", "bact", "bact_a", "ylc", "ylc_a", "sperm", "sperm_a"]
    return new_data
```

```python
get_norm_target_data(sillyfilepath)
```
